```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

LINE Chart

```python
plt.plot(x,y)
plt.title ("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar Chart

```python
plt.bar(x,y,color=['red','orange'])
plt.title ("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

scatter plot

```python
a=(1,9,3,8,5)
b=(11,12,13,15,24)

plt.scatter(a,b)
plt.title ("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
